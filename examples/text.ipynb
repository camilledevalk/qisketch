# Qisketch text example

This Jupyter notebook shows how to convert text to a quantum circuit using Qisketch.

```python
from src.qisketch import (
    add_bleed,
    create_qc_from_pixels,
    char_to_pixels,
)
import numpy as np

# Define the text
text = "Quantum Rules"

# Convert the text to pixels
pixels = char_to_pixels(text, size=2)

# Add the bleed
pixels = add_bleed(pixels)
```

```python
# Create the quantum circuit
qc = create_qc_from_pixels(pixels, control_probability=0.3, seed=1)

# Draw the quantum circuit
qc.draw(
    output="mpl",
    fold=0,
    plot_barriers=False,
    filename=f"examples/outputs/{text}-circuit.png",
)
```
